## Extraction of Data from S3

This notebook shows steps to create Dataframe extracted from data stored in AWS S3. There are two ways to establish access to S3: IAM roles and access keys. We have used access keys for this purpose. Below is the code:

```
import urllib
ACCESS_KEY = "AWS ACCESS KEY"
SECRET_KEY = "AWS SECRET KEY"
ENCODED_SECRET_KEY = urllib.parse.quote(SECRET_KEY, "")
AWS_BUCKET_NAME = "sample-clickstream"
MOUNT_NAME = "sample-data"
dbutils.fs.mount("s3n://%s:%s@%s" % (ACCESS_KEY, ENCODED_SECRET_KEY, AWS_BUCKET_NAME), "/mnt/%s" % MOUNT_NAME)
```

```
display(dbutils.fs.ls("mnt/sample-data"))
```

```
# Create a dataframe with the anticipated schema
df_clickstream = (sqlContext.read.format('com.databricks.spark.csv')
  .options(
    header='true',  # first line is the header
    delimiter=',',  # tab delimiter
    mode='PERMISSIVE',
    infer_schema='true',  # infer data types (e.g., int, string) from values
  )
  .load("/mnt/sample-data/clickstream-sample")
)
```
